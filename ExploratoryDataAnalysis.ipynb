```python
# The average yearly spending by gender
query = """
SELECT Gender, SUM(YearlyAmountSpent) AS TotalSpent
FROM customers_data
GROUP BY Gender;
"""
pd.read_sql(query, engine)
```

```python
# profession spends the most on average
query = """
SELECT Profession, AVG(YearlyAmountSpent) AS AvgSpent
FROM customers_data
GROUP BY Profession
ORDER BY AvgSpent DESC;
"""
pd.read_sql(query, engine)
```

```python
# The age distribution of customers
query = """
SELECT 
  CASE 
    WHEN Age < 25 THEN 'Under 25'
    WHEN Age BETWEEN 25 AND 34 THEN '25-34'
    WHEN Age BETWEEN 35 AND 44 THEN '35-44'
    WHEN Age BETWEEN 45 AND 54 THEN '45-54'
    ELSE '55+' 
  END AS AgeGroup,
  COUNT(*) AS CustomerCount
FROM customers_data
GROUP BY AgeGroup;
"""
pd.read_sql(query, engine)
```

```python
#the most preferred brand by gender
query = """
SELECT Gender, PreferredBrand, COUNT(*) AS PreferenceCount
FROM customers_data
GROUP BY Gender, PreferredBrand
ORDER BY Gender, PreferenceCount DESC;
"""
pd.read_sql(query, engine)
```

```python
# average loyalty score by membership tier
query = """
SELECT MembershipTier, AVG(LoyaltyScore) AS AvgLoyalty
FROM customers_data
WHERE MembershipTier IS NOT NULL
GROUP BY MembershipTier
ORDER BY AvgLoyalty DESC;
"""
pd.read_sql(query, engine)
```

```python
query = """
SELECT MembershipTier, SUM(YearlyAmountSpent) AS AmountByMembership
FROM customers_data
WHERE MembershipTier IS NOT NULL
GROUP BY MembershipTier
ORDER BY AmountByMembership DESC;
"""
pd.read_sql(query, engine)
```

```python
query = """
SELECT Category,
       SUM(Quantity * (UnitPrice - Discount)) AS TotalRevenue
FROM orders_data
GROUP BY Category
ORDER BY TotalRevenue DESC;
"""
pd.read_sql(query, engine)
```

```python
# Top-Selling Categories
query = """
SELECT Category, 
       SUM(Quantity) AS TotalUnitsSold
FROM orders_data
GROUP BY Category
ORDER BY TotalUnitsSold DESC;
"""
pd.read_sql(query, engine)
```

```python
#  Total Revenue by Sub-Category
query = """
SELECT SubCategory, 
       SUM(Quantity * (UnitPrice - Discount)) AS TotalRevenue
FROM orders_data
GROUP BY SubCategory
ORDER BY TotalRevenue DESC;
"""
pd.read_sql(query, engine)
```

```python
# Most Profitable Brands
query = """
SELECT Brand, 
       SUM(Profit) AS TotalProfit
FROM orders_data
GROUP BY Brand
ORDER BY TotalProfit DESC
LIMIT 10;
"""
pd.read_sql(query, engine)
```

```python
# Sales Trend Over Time
query="""
SELECT 
    DATE(OrderDate) AS OrderDay,
    SUM(Quantity * (UnitPrice - Discount)) AS DailyRevenue
FROM orders_data
GROUP BY OrderDay
ORDER BY OrderDay;
"""
pd.read_sql(query, engine)
```

```python
# Average Order Value by Region
query = """
SELECT Region, 
       AVG(Quantity * (UnitPrice - Discount)) AS AvgOrderValue
FROM orders_data
GROUP BY Region
ORDER BY AvgOrderValue DESC;
"""
pd.read_sql(query, engine)
```

```python
# Preferred Payment Methods
query="""
SELECT PaymentMethod, 
       COUNT(*) AS OrdersCount
FROM orders_data
GROUP BY PaymentMethod
ORDER BY OrdersCount DESC;
"""
pd.read_sql(query, engine)
```

```python
query="""
SELECT DATEDIFF(ShipDate, OrderDate) AS ShippingDays
FROM orders_data
WHERE ShipDate IS NOT NULL;
"""
pd.read_sql(query, engine)
```

```python
# top 10 customers
query="""
SELECT CustomerID, SUM(Profit) AS TotalProfit
FROM orders_data
GROUP BY CustomerID
ORDER BY TotalProfit DESC
LIMIT 10;
"""
pd.read_sql(query, engine)
```
